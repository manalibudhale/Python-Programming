```python
#Checking outliers using boxplot
sum=1
plt.figure(figsize=(15,10))
for i in ['Unit price', 'Quantity', 'Tax 5%', 'Total','cogs', 'gross income', 'Rating']:
    plt.subplot(3,3,sum)
    plt.boxplot(df[i])
    sum=sum+1
    plt.title(i)
```

```python
#Observation - Tax 5%, Total, cogs and gross income columns have outliers
#Also, dropping column branch as it is giving same information as City
```

```python
df.drop(columns=['gross income','cogs','Branch'],inplace=True)
```

```python
df['Total'].agg(['sum','mean','median','std'])
```

```python
plt.figure(figsize=(15,10),facecolor='pink')
axis=plt.subplot(2,1,1)
axis.set_facecolor('gray')
sns.barplot(data=df, x='City', y='Total', palette='bright', capsize=.01)
plt.title('Average sale for each City',fontdict={'fontsize': 15,'fontweight': 'bold','color': 'black'})
axis1=plt.subplot(2,1,2)
axis1.set_facecolor('gray')
sns.barplot(data=df, x='City', y='Total', estimator='sum', palette='bright', capsize=.01)
plt.title('Total sale for each City',fontdict={'fontsize': 15,'fontweight': 'bold','color': 'black'});
```

```python
#Observation - Naypyitaw city has the highest sale
#Let's find the highest sale between male and female and also the type of products
```

```python
sns.barplot(data=df, x='Gender', y='Total', estimator='sum')
plt.xlabel('Gender')
plt.ylabel('Cost')
plt.title('Total sales as per Gender')
plt.show()
```

```python
#So, from above diagram we can derive that Female spend more on Products than Male
```

```python
p=df['Product line'].value_counts()
p
```

```python
plt.figure(facecolor='pink')
clrs = ["red", "hotpink", "lime", "yellow",'cyan','white']
plt.pie(p.values,labels=p.index,autopct='%.2f%%',colors=clrs,explode=[0.2,0,0,0,0,0],wedgeprops={'linewidth': 2.0, 'edgecolor': 'black'})
plt.title('Most popular product',color='black',fontsize=15,fontweight='bold');
```

```python
#Most selling product
df.groupby(['Product line'])['Total'].agg(['sum'])
```

```python
plt.figure(figsize=(10,5))
x = plt.axes()
sns.barplot(x='Product line',y='Total',data=df,estimator='sum',palette='bright',facecolor='silver')
plt.xticks(rotation=45);
```

```python
#Above graph depicts that Food and beverages was the most selling product whereas Health and beauty was least selling product
```
